<p style='text-align: center;'> <font size = '5;' >  <b> Abstract  </b> </font> </p> 

<p style='text-align: justify;'> The profiles of aerosol extinction coefficient simulated by two models during the AeroCom 2019 project  is compared with the CALIOP (Cloud-Aerosol Lidar with Orthogonal Polarization) retrievals for the Arctic region. The models considered for this study are Geophysical Fluid Dynamics Laboratory (GFDL) and Spectral Radiation-Transport Model for Aerosol Species (SPRINTARS). Monthly averaged globally gridded model and satellite data are considered for this purpose for the year 2010 over the Arctic region. The seasonal variation of the mean extinction height diagnostics ($Z_\alpha$) is studied for the CALIOP and model data for altitudes up to 10 km. The models are found to overestimate the $Z_\alpha$ values in annual and seasonal scales. However, the seasonal variation shown by models is at par with the CALIOP retrievals.</p>   <br> <br> 

 <font size = '5;' ><b>1. Introduction  </b> </font>

<p style='text-align: justify;'> Aerosols are solid or liquid particles suspended in the atmosphere. They are generally very tiny for naked human eye but are mighty when it comes to the Earth radiation budget. They not only scatter and absorb the incoming solar and outgoing terrestrial radiation (direct forcing) but also act as condensation nuclei forming clouds and thereby effecting the radiation budget (indirect forcing) (Myhre et al., 2014). The vertical distribution of aerosols highly impacts the transport of (shortwave and longwave) radiation through the atmosphere as well as the formation of clouds and precipitation patterns (Mishra et al., 2015). It also reflects aerosol lifetime (higher up aerosols indicating a longer lifetime). Thus when it comes to the assessment of climate relevant parameters such as the anthropogenic aerosol radiative forcing, which is among the highest uncertainties in our climate models (Boucher et al., 2013), the vertical distribtuion of aerosols plays a crucial role. Therefore, it is imortant for the aerosol models to correctly simulate the vertical distribution of aerosols.</p> 

<p style='text-align: justify;'> CALIOP (Cloud-Aerosol Lidar with Orthogonal Polarization) is a spaceborne lidar on board CALIPSO (Cloud-Aerosol Lidar and Infrared Pathfinder Satellite Observation) satellite which measures the aerosol extinction profiles in high spatial resolution. In this report, the CALIOP retrievals are used to evaluate the aerosol (climate) models from the AeroCom 2019 project. The models considered for the evaluation are the Geophysical Fluid Dynamics Laboratory (GFDL) and Spectral Radiation-Transport Model for Aerosol Species (SPRINTARS). The mean aerosol extinction height diagnostic ($Z_\alpha$) is used in this work as a prognostic variable to assess model performance with respect to the vertical distribution of aerosols (koffi et al., 2012; 2016). It is defined by the formula </p>

\begin{equation*}
Z_\alpha =\frac{\sum_{i=0}^n \beta_i * Z_i}{\sum_{i=0}^n \beta_i} \tag{1}
\end{equation*}

<p style='text-align: justify;'> Where, $\beta_i$ and $Z_i$ are the aerosol extinction coefficient (in $km^{-1}$) and altitude at level ùëñ respectively. In this work, the $Z_\alpha$ [$eq^n(1)$] is computed for altitudes up to 10 km for annual and seasonal scales. </p>

 <font size = '5;' >  <b>2. Data and Methodology  </b> </font>

<p style='text-align: justify;'>The CALIPSO level 3 aerosol profile product reports monthly mean profiles of aerosol optical properties on a uniform spatial grid for different sky condition and lighting condition. This is a tropospheric product that reports averaged values at altitudes below 12km. All level 3 parameters are derived from the version 3 CALIOP level 2 aerosol profile product and are quality screened prior to averaging. The CALIPSO data used for this work is for cloud free and day conditions. The AeroCom 2019 models used for this study are GFDL and MIROC-SPRINTARS. The data details including the resolution are shown in table 1. Both the satellite and model data are monthly data. The data selected for this work is from the year 2010.</p> 

<p style='text-align: center;'><b>Table 1:</b> Satellite and AeroCom model data details</p> 

| Data | Experiment name (model) <br> or data type (observation)| Spatial resolution <br> (lon x lat x levels)|
| --- | --- | ---|
| CALIOP (observation)| level 3 monthly daytime cloud free   | $5^o$ X $2^o$ X $208$| 
|GFDL (model)| GFDL-AM4-met2010_AP3-CTRL |$1.25^o$ X $1^o$ X $33$|
|MIROC-SPRINTARS (model)|MIROC-SPRINTARS_AP3-CTRL|$0.56^o$ X $0.56^o$ X $57$|

<p style='text-align: justify;'>The primary objective of this work is to asses the aerosol vertical distribution simulations from the models. Since most of the aerosols lie below tropopause, the altitude limit is set to 10 kilometers. The regions with latitude greater than $70^o$N are selected for all the data analysis so as to approximate the Arctic. The area of lon-lat grid box is different for different latitudes and tends to decrease as we move towards the poles. In order to compensate this effect, area weighted average of extinction profile is opted for the data analysis. Both the models have hybrid sigma-pressure vertical coordinate system. The atlitudes of each vertical level is calculated by using the layer thickness for each level and the surface elevations both of which are reported in a separate file inside the model path. The altitude of each level is determined by adding all the layer thicknesses to the suface elevations from bottom to top. The next section presents the results and the associated python codes.</p>

 <font size = '4.5;' ><b>3.1. Profiles of extinction coefficient </b> </font>

 <font size = '3.5;' ><b>3.1.1. Annual extinction profiles </b> </font>

<p style='text-align: justify;'>The annual mean extinction profiles are computed in this section for CALIOP and model data. Before averaging, the data is cropped for latitudes greater than $70^o$N and for altitudes smaller than 10 kilometers as mentioned in the Section 2. Moreover, the normalised extinction profiles and the annual mean extinction height $Z_\alpha$ [from $eq^n(1)$] are also shown in this section.</p>  

```python
# Estimate the CALIOP absolute and normalised extinction profile
_dataset = xr.open_dataset(CALIOP_path+caliop_id)
_extinction_caliop = _dataset['ec532aer']
extinction_caliop_10km = _extinction_caliop.sel(latitude=slice(70,90),level=slice(0,10))
profile_caliop_height_10km = _dataset['level'].sel(level=slice(0,10))
profile_caliop_extinction_10km = extinction_caliop_10km.mean(dim = ['time','longitude','latitude'],skipna = 'True', keep_attrs = 'True')
# Normalise the extinction coefficient 
norm_profile_caliop_extinction_10km= normalised_profile(profile_caliop_extinction_10km)
# Compute the mean extinction height using the function "mean_extinction_height" 
Zo=mean_extinction_height(profile_caliop_extinction_10km,profile_caliop_height_10km)
```

<p style='text-align: justify;'>The model data needed a lot of processing to produce extinction profiles comparable with the CALIOP profiles. This includes computing the atlitudes for the model vertical levels and performing area weighted averaging. Thus to simplify the code and plot the model profiles using one cell (so as to minimise the JupyterHub server workspace), many functions are used in the code below.</p> 

```python
# Create a figure of desired size
fig,ax = plt.subplots(1,2,figsize=[12,6])
plt.rcParams.update({'font.size': 12})
axs = ax.flatten()
fig.suptitle('Annual mean profiles of extinction coefficient (Model vs Observation)' ,fontsize=16)

# First plotting the CALIOP profile
axs[0].plot(profile_caliop_extinction_10km,profile_caliop_height_10km,label='CALIOP satellite data'+ '\n'+' (Z = '+np.array2string(Zo)+')')
axs[1].plot(norm_profile_caliop_extinction_10km,profile_caliop_height_10km,label='CALIOP satellite data')


# Plotting the model extinction profiles using a loop 
for i in range(len(modelid)):
    path_to_model_data_folder = aerocom_path + modelid[i]
    profile_model_extinction_10km, profile_model_height_10km = compute_model_profile_10km(path_to_model_data_folder)

#Calculate the normalised extinction coefficient for models
    norm_profile_model_extinction_10km= normalised_profile(profile_model_extinction_10km)
    
#Calculate the mean aerosol extinction height for model (Zm)     
    Zm=mean_extinction_height(profile_model_extinction_10km,profile_model_height_10km)
    
    axs[0].plot(profile_model_extinction_10km,profile_model_height_10km, '-',label=modelid[i]+ '\n'+' (Z = '+ np.array2string(Zm)+')')
    axs[1].plot(norm_profile_model_extinction_10km,profile_model_height_10km,'-',label=modelid[i])
    set_axis_profiles_model_caliop(axs)
```

<b> Figure 1:</b> Absolute and normalised profiles of annual mean extinction coefficient (in $km^{-1}$) for the year 2010 averaged over the Arctic

<p style='text-align: justify;'>The figure 1 shows the absolute and normalised aerosol extinction profiles for CALIOP and model data averaged over the year 2010 and the Arctic region. It can be seen that both the models overestimate the absolute value of annual mean aerosol extinction profile for most part of the atmosphere with the MIROC-SPLINTARS model performing better than the GFDL model. Generally, the model parameters starting from the grid resolution to the parametrizations used to approximate the physical processes, vary from one model to the other. So it is better to check if the models are able to simulate the vertical variations rather than the magnitude only. To do so, the extinction coefficients are normalised to one and are shown in the right hand side of the figure 1. It can be seen that the models are quite accurate in simulating the vertical variation of extinction coefficient in the lower atmosphere below 1 km. However, for atlitudes above 2 kms, both the models fail miserably. The seasonal profiles are presented in the next section.</p> 

<font size = '3.5;'><b>3.1.2 Seasonal extinction profiles </b> </font>

<p style='text-align: justify;'>The seasonal mean extinction profiles are computed in this section for CALIOP and model data. The standard seasons such as Dec-Jan-Feb, Mar-Apr-May, Jun-July-Aug and Sept-Oct-Nov are considered for this study. The rest of the calculations are similar to the one described in the Section 3.1.1 with the procedure being repeated for all the seasons.</p>  

```python
seasonid=[-1,3,6,9]
seasonname=['D-J-F','M-A-M','J-J-A','S-O-N']

# Create a figure of desired size and update the parameters
plt.rcParams.update({'font.size': 12,'font.sans-serif': 'Bold', 'axes.linewidth': 2})
fig,ax = plt.subplots(2,2,figsize=[10.5,10.5])
axs = ax.flatten()
get_done_titles(axs)
plt.text(-0.046,-3,'Extinction Coefficient (km-1)',fontsize=14, fontweight='bold')
plt.text(-0.093,13,'Height (km)',fontsize=14, rotation=90,fontweight='bold')
plt.text(-0.08,25,'Seasonal mean profiles of extinction Coefficient (Model vs Observation)',fontsize=16)

# Creating a dummy null array to store all the mean extinction height values to be used in the next section
ZZ=np.zeros([3,4])

# Looping over all the seasons
for j in range(len(seasonid)):
    # First computing the extinction profile and mean height for CALIOP data
    s_profile_caliop_extinction_10km=compute_seasonal_caliop_profile_10km(extinction_caliop_10km,seasonid[j])
    ZZ[0,j]=mean_extinction_height(s_profile_caliop_extinction_10km,profile_caliop_height_10km)

    if j==0:
        axs[j].plot(s_profile_caliop_extinction_10km,profile_caliop_height_10km, '-',label='CALIOP satellite data' + '\n'+' (Z = '+ np.array2string(ZZ[0,j])+')')    
    else:
        axs[j].plot(s_profile_caliop_extinction_10km,profile_caliop_height_10km, '-',label='Z = '+ np.array2string(ZZ[0,j]))
    for i in range(len(modelid)):
        # Computing the extinction profile and mean height for model data
        path_to_model_data_folder = aerocom_path + modelid[i]
        s_profile_model_extinction_10km, s_profile_model_height_10km = compute_seasonal_model_profile_10km(path_to_model_data_folder,seasonid[j])
        s_Zm=mean_extinction_height(s_profile_model_extinction_10km,s_profile_model_height_10km)
        ZZ[i+1,j]=s_Zm
        if j==0:
            axs[j].plot(s_profile_model_extinction_10km,s_profile_model_height_10km, '-',
                        label=modelid[i]+ '\n'+' (Z = '+ np.array2string(s_Zm)+')')
        else:
            axs[j].plot(s_profile_model_extinction_10km,s_profile_model_height_10km, '-',
                        label='Z = '+ np.array2string(s_Zm))
        axs[j].legend()
        axs[j].set_xlim(-0.005,0.06)
```

<b>Figure 2:</b> Seasonal mean extinction coefficients in km-1 for model outputs and CALIOP data for the year 2010 

<p style='text-align: justify;'>The figure 2 shows the seasonal extinction profiles for the CALIOP and model data. The CALIOP extinction profile shows relatively higher values close to the surface in D-J-F and S-O-N (winter and autumn months). Such higher values close to the surface may occur because of accumulation of aerosols due to the stable atmosphere (temperature inversion) and because of lesser frontal activities (less precipitation implies less wet deposition). When it comes to models, the GFDL model performs better than MIROC-SPLINTARS model in simulating the lower level extinction profile in D-J-F and S-O-N months, whereas the MIROC-SPLINTARS performs better in rest of the seasons.</p>

This section presents the seasonal variation of mean extinction height calculated from the $eq^n(1)$ for altitudes up to 10 km. 

```python
# Create a figure of desired size
fig,ax = plt.subplots(figsize=[10,5])
id=[0,1,2,3]
plt.rcParams.update({'font.size': 12})

# Plot the mean extinction height (already estimated in Section 3.1.2) 
ax.plot(id,ZZ[0,:],label='CALIOP satellite data')
ax.plot(id,ZZ[1,:],label='GFDL-AM4-met2010_AP3-CTRL')
ax.plot(id,ZZ[2,:],label='MIROC-SPRINTARS_AP3-CTRL')
set_axes_extinction_height(ax,id,seasonname)
```

<b>Figure 3:</b> CALIOP and modeled mean extinction height (in km) for different seasons over the Arctic

<p style='text-align: justify;'>As seen in the figure 3, the mean extinction height is comparatively less in D-J-F and S-O-N months in both the models and observations. This may occur because the aerosols tend to acumulate in the lower troposphere close to the surface due to stable atmospheric conditions and comparatively less surface fluxes (low boundary layer height). The models overestimate the mean extinction height in all the season. However, the GFDL model is capable of simulating the seasonal variation of mean extinction height. MIROC-SPLINTARS performs relatively better in simulating the magnitude of seasonal mean extinction height. <font size = '5;' ><b>4. Conclusion and outlook  </b> </font>

<p style='text-align: justify;'>The aerosol extinction profiles simulated by the AeroCom models are compared with the CALIOP satellite observations for the year 2010 over the Artic region. The profiles are compared for annual and seasonal time scales. Mean extinction height is computed for altitudes up to 10 km and compared for annual and seasonal scales. MIROC-SPLINTARS model performs better in simulating the annual mean aerosol extinction profile compared to the GFDL model. The CALIOP extinction profile shows relatively higher values close to the surface in D-J-F and S-O-N (winter and autumn months). The models overestimate the mean extinction height in all the season but are capable of simulating the seasonal variation of mean extinction height. MIROC-SPLINTARS performs relatively better in simulating the magnitude of seasonal mean extinction height but the seasonal variation is better simulated by the GFDL model. At the current stage, it is hard to say which model performs better and why? Due to time constraint, the data analysis is limited to two models and the time period considered is one year, which are not enough to come to any general conclusion. A detailed analysis of different types of model simulations needs to be done for a longer time period to properly document the model performances relative to each other.</p>
